```
from sklearn import datasets
import matplotlib.pyplot as plt
import math

iris = datasets.load_iris()

# Enregistrement dans la variable X des longueurs des pétales
# et dans la variable Y des largeurs des pétales
X = iris.data[:, 2]
Y = iris.data[:, 3]

# Les masques contenus dans le tableau iris.target vont être utlisés pour donner une couleur à chaque espèce d'iris
Masque_Couleur = iris.target

plt.scatter(X, Y, c=Masque_Couleur)
plt.xlabel('Longueur des pétales (cm)')
plt.ylabel('Largeur des pétales (cm)')
plt.title('Distribution des dimensions des pétales')
```
